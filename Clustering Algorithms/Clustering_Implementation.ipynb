Import all the required libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

Generate the dataset

```python
X, _ = make_blobs(n_samples=500, centers = 3, cluster_std=0.60, random_state=44)
plt.scatter(X[:, 0], X[:, 1], s=50)
plt.title('Unlabeled Synthetic Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
```

```python
X
```

K Means Clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=44)
label_kmeans = kmeans.fit_predict(X)
```

```python
X
```

```python
label_kmeans
```

```python
kmeans.cluster_centers_
```

```python
plt.scatter(X[:, 0], X[:, 1], c=label_kmeans, s=50)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='red', marker='x', s=200, label='centroids')
plt.legend()
plt.show()
```

Create the Elbow Graph

```python
wcss = []
k_range = range(1, 11)

for k in k_range:
  kmeans = KMeans(n_clusters=k, random_state=44)
  kmeans.fit(X)
  wcss.append(kmeans.inertia_)

plt.figure(figsize=(8,5))
plt.plot(k_range, wcss, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of clusters (K)')
plt.ylabel('Within Cluster Sum Of Squares (WCSS)')
plt.show()
```

Hierarchial Clustering

```python
from sklearn.cluster import AgglomerativeClustering

hierachial_model = AgglomerativeClustering(n_clusters=3)
labels_hierarchial = hierachial_model.fit_predict(X)

plt.scatter(X[:, 0], X[:, 1], c=labels_hierarchial, s=50)
plt.title('Hierarchial Clustering Result')
plt.show()
```

DBSCAN Implementation

```python
from sklearn.cluster import DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=labels_dbscan, s=50)
plt.title("DBSCAN clustering algorithm")
plt.show()
```
