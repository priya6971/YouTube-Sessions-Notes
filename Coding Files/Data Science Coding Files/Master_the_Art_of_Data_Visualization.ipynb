**Import all the required libraries which are useful for the Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Inference1: Positive Relationship between X & Y

Inference 2: Negative Relationship between X & Z

```python
X = [1, 3, 7]
Y = [1, 5, 8]
Z = [7, 2, 1]

plt.title("Visualization of the data")
plt.xlabel("X")
plt.ylabel("Y & Z")

plt.plot(X, Y)
plt.plot(X, Z)

## dealing with more than one representation in a single graph
plt.legend(["Plot b/w X&Y", "Plot b/w X&Z"])

## save the visualizations
plt.savefig("Visualization1.png")

plt.show()
```

Inference: 0s in the Glucose and BP doesn't makes sense in the dataset. So, we need to do the data imputation for all the 0's available in every feature.

PairPlot - Wrt every independent features what are the inferences you are getting?

```python
plotImg = sns.pairplot(data)
plotImg.savefig("PairPlot.png")
```

Distribution of the features: Symmetric and Non-Symmetric Distribution

Symmetric - Mean

Non-Symmetric - Median

Why Median is more robust to outliers?

Inference: Non-symmetric Distribution - Right Skewed(Median)

```python
sns.distplot(data.Insulin)
```

Outliers Detection: BoxPlot

Inference: Insulin feature is having the maximum outliers in the given dataset

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.boxplot(data=data, ax=ax)
plt.savefig("boxplot.png")
```

Inference: Select the important/crucial features in our dataset which is responsible to make correct predictions of the trained model.

```python
plt.figure(figsize=(15, 15))
ax = sns.heatmap(data.corr(), annot=True)
plt.savefig("Correlation Coefficient Heatmap.png")
plt.show()
```
